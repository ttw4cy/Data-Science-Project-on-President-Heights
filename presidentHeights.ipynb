```python
# Importing numpy and pandas
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("president_heights.csv")
print(data.head())
```

```python
height = np.array(data["height(cm)"])
print(height)
```

```python
print("Média das altura =", height.mean())
print("Desvio padrão de altura =", height.std())
print("Minima altura =", height.min())
print("Maxima altura =", height.max())
```

```python
print("25o percentil =", np.percentile(height, 25))
print("Mediana =", np.median(height))
print("75o percentil =", np.percentile(height, 75))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.hist(height)
plt.title("Distribuição de altura de presidentes do EUA")
plt.xlabel("Altura(cm)")
plt.ylabel("Quantidade")
plt.show()
```
